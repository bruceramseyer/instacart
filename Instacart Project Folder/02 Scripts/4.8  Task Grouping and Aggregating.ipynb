# 4.8 Task: Grouping and Aggregating

```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

This script contains the following points:
1. Importing orders and products data
2. Grouping and aggregrating values
3. Creating user-defined field for Loyalty_flag, spending_flag, and frequency_flag

```python
path=r'C:\Users\dell\Desktop\Career Foundry\0423_Instacart Analysis'
```

```python
# Import pickle for order_products_combined data set
ords_prods_merge=pd.read_pickle(os.path.join(path,'02 Data','Prepared Data','ord_prod_newvar.pkl'))
```

```python
# Group and mean of department_id

ords_prods_merge.groupby('department_id').agg({'order_number':['mean']})
```

# Analyzing the mean for the entire data set against the sub_set, additional department_ids now appear. Most of the averages decreased from the values in the subset with the exception of department_id 17; which increased from ~11.29 to ~15.69.

```python
# Creating max_order field via max aggregate for Loyalty Flag

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

# Setting Loyalty Flags

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

```python
ords_prods_merge.head(10)
```

```python
# Grouping loyalty_flag with mean, min, max prices

ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
# Creating mean_spend field via mean aggregate for Spending Flag

ords_prods_merge['mean_spend'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge[['user_id','mean_spend','order_number']].head(15)
```

# Setting Spending Flags

```python
ords_prods_merge.loc[ords_prods_merge['mean_spend'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_spend'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

```python
# Creating median_spend field via mean aggregate for Frequency Flag

ords_prods_merge['median_freq'] = ords_prods_merge.groupby(['user_id'])['#ofdays_since_last_order'].transform(np.median)
```

```python
ords_prods_merge[['user_id','median_freq','
                  ']].head(15)
```

User_id returns a median_freq of 8, but median appears to be 7

# Setting Frequency Flags

```python
ords_prods_merge.loc[ords_prods_merge['median_freq'] > 20, 'frequency_flag'] = 'Non-Frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_freq'] <= 20) & (ords_prods_merge['median_freq'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_freq'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```
