```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

This script contains the following points:
1. Importing of orders and products data
2. Merging of orders and products data

```python
path=r'C:\Users\dell\Desktop\Career Foundry\0423_Instacart Analysis'
```

```python
# Import pickle for order_products_combined data set
df_orders_prods_combined=pd.read_pickle(os.path.join(path,'02 Data','Prepared Data','orders_products_combined.pkl'))
```

```python
df_orders_prods_combined.shape
```

```python
pvars_list=['product_id','product_name','aisle_id','department_id','prices']
```

```python
# Import data set products.csv
df_prods=pd.read_csv(os.path.join(path,'02 Data','Prepared Data','products_checked.csv'),usecols=pvars_list)
```

```python
df_prods.head()
```

```python
df_orders_prods_combined.head()
```

```python
# Test merge without overwrite

pd.merge(df_orders_prods_combined,df_prods, on = ['product_id'], indicator = True)
```

```python
# Merge of orders and products dataframes with indicator

df_merged_indicator = df_orders_prods_combined.merge(df_prods, on = ['product_id'], indicator = True)
```

```python
df_merged_indicator['_merge'].value_counts()
```

```python
# Final merge without indicator for exporting

df_orders_prods_merged = df_orders_prods_combined.merge(df_prods, on = ['product_id'])
```

```python
# Export data to pkl

df_orders_prods_merged.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_prods_merged.pkl'))
```
